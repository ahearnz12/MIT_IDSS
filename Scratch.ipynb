```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the CSV file into a DataFrame
df = pd.read_csv('/Users/alexanderhearnz/codebase/MIT_IDMSS/data/prog_book.csv')

# Histogram
plt.figure(figsize=(10, 6))
sns.histplot(df['Reviews'], kde=False)
plt.title('Histogram of Book Reviews')
plt.xlabel('Number of Reviews')
plt.ylabel('Frequency')
plt.show()

# KDE Plot
plt.figure(figsize=(10, 6))
sns.kdeplot(df['Reviews'], shade=True)
plt.title('KDE Plot of Book Reviews')
plt.xlabel('Number of Reviews')
plt.ylabel('Density')
plt.show()

# Combined Histogram and KDE Plot
plt.figure(figsize=(10, 6))
sns.histplot(df['Reviews'], kde=True)
plt.title('Histogram and KDE Plot of Book Reviews')
plt.xlabel('Number of Reviews')
plt.ylabel('Frequency')
plt.show()

# Box Plot
plt.figure(figsize=(10, 6))
sns.boxplot(y=df['Reviews'])
plt.title('Box Plot of Book Reviews')
plt.ylabel('Number of Reviews')
plt.show()

# Violin Plot
plt.figure(figsize=(10, 6))
sns.violinplot(y=df['Reviews'])
plt.title('Violin Plot of Book Reviews')
plt.ylabel('Number of Reviews')
plt.show()
```
